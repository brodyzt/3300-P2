```python
df = df.dropna()
bar_categories = ["Genre","Platform"]
axis_categories = ["Genre","Platform","Year"]

with open('stacked_bar_data.json', 'w') as outfile: 
    json_data = dict()
    
    for b_category in bar_categories:
        for x_category in axis_categories:
            data_per_b_x = []
            for x_val in df[x_category].unique():
                value_sum = 0
                b_counts = []
                for b_val in df[b_category].unique():
#                     print(b_val)
                    cell_count = len(df[(df[b_category] == b_val) & (df[x_category] == x_val)])
                    b_counts.append([b_val, cell_count])
                    value_sum += cell_count
            
                for d in b_counts:
                    if value_sum != 0: d[1] /= value_sum
                
#                 b_counts = sorted(b_counts, key=lambda x: x[1], reverse=True)
                data_per_b_x.append([x_val,b_counts])
                    
                
#                 b_counts = sorted(b_counts, key=lambda x: x[1], reverse=True)
#                 temp_data.append([b_category, b_counts])
            
            if x_category == "Year":
                data_per_b_x = sorted(data_per_b_x, key=lambda x: x[0])
                                      
            key = b_category + "_" + x_category    
            json_data[key] = data_per_b_x
            
#             print(json_data)
    
    json.dump(json_data, outfile)
```
