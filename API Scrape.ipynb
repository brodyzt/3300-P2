```python
with open("game_api_data.json", "w") as game_file:
    df = pd.read_csv("games.csv")
    
    try:
        data_dict = pickle.load( open( "data.pkl", "rb" ) )
        if "failed" not in data_dict:
            data_dict["failed"] = {}
    except:
        data_dict = dict()
    
    i = 0
    for game_title in list(df[df["Rank"] < 2000]["Name"]):
        print(str(i / 2000.0) + "% done")
        if game_title in data_dict or game_title in data_dict["failed"]:
            print(game_title + " has already been scraped")
        else:
            print("Retrieving " + game_title)
             # get most likely game ID
            url = "https://api-v3.igdb.com/games"
            fields = "age_ratings,aggregated_rating,aggregated_rating_count,artworks,bundles,category,collection,cover,created_at,dlcs,expansions,external_games,first_release_date,follows,franchise,franchises,game_engines,game_modes,genres,hypes,involved_companies,keywords,multiplayer_modes,name,parent_game,platforms,player_perspectives,popularity,pulse_count,rating,rating_count,release_dates,screenshots,similar_games,slug,standalone_expansions,status,storyline,summary,tags,total_rating,total_rating_count,updated_at,url,version_parent,version_title,videos,websites"
            querystring = {"search":game_title,"fields":fields}

            payload = ""
            headers = {
                'user-key': "62d64294c2ed1551b9e06030e90ad8bc"
                }

            response = requests.request("GET", url, data=payload, headers=headers, params=querystring)
            
            try:
#                 print(response.text)
                game_data = response.json()[0]
                game_id = game_data["id"]

                data_dict[game_title] = game_data

                pickle.dump(data_dict, open( "data.pkl", "wb" ) )
            except Exception as e:
                data_dict["failed"][game_title] = 0
                print("Failed on this data:")
                print(e)
                print(response.json())
                
        i += 1
                
    json.dump(data_dict, game_file)
```
